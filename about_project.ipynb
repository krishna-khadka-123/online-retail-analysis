# Online Retail Data Analysis

This project analyzes an **online retail dataset** to uncover key insights about sales, products, and customer behavior. The analysis was performed using Python in Jupyter Notebook with a focus on both exploratory data analysis (EDA) and business-relevant metrics.

## Dataset
- File: `online_retail_dataset.csv`
- Contains transactions, product details, customer country, and sales information.

## Analysis Performed
The following steps were carried out in this project:

1. **Feature Engineering** – Created new features to simplify analysis and extract meaningful insights.  
2. **Statistical Summary** – Generated basic statistics to understand distributions, outliers, and trends in the dataset.  
3. **Gross Sales by Month (2025)** – Analyzed monthly sales trends for the year 2025.  
4. **Correlation Heatmap** – Visualized the correlation between numeric variables to identify relationships.  
5. **Top 10 Selling Products** – Identified the most frequently sold products.  
6. **Monthly Revenue for All Years** – Tracked revenue trends across different months over multiple years.  
7. **Highest Revenue Products** – Determined which products contributed the most to total revenue.  
8. **Top 10 Products by Revenue** – Focused on the top revenue-generating products for deeper insights.  
9. **Product Category Revenue Analysis** – Found which product categories generate the most revenue.  
10. **Country Sales Analysis** – Examined sales distribution across different countries.  
11. **Most Used Payment Method** – Analyzed which payment methods are most popular among customers.

## Tools and Libraries
- **Python**  
- **Pandas** for data manipulation  
- **NumPy** for numerical computations  
- **Matplotlib & Seaborn** for visualization  
- **Jupyter Notebook** for interactive analysis  

## Key Insights
- Monthly and yearly sales trends reveal seasonal patterns in customer purchases.  
- A small number of products generate the majority of revenue, highlighting best-sellers.  
- Certain product categories consistently outperform others in terms of revenue.  
- Most sales come from a few key countries, suggesting potential focus markets.  
- Credit/Debit card or the primary payment method dominates customer transactions.  

## Conclusion
This project demonstrates how data analysis can provide actionable insights for business strategy, inventory planning, and sales optimization in the e-commerce sector. The visualizations and summaries make it easy to identify trends, top-performing products, and customer behavior patterns.
